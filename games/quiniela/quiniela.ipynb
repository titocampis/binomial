```python
# Media de datos
med_people = []
med_percent = []
med_unitary = []

med_apuestas = round(np.mean(apuestas))
for i in range(len(people)):
  med_people.append(np.mean(people[i]))
  med_percent.append(np.mean(percent[i]))
  med_unitary.append(np.mean(pr_unit[i]))

print("Media de participaci√≥n (personas)):", med_apuestas)
print()
print("Media de personas que aciertan (10, 11, 12, 13 y 14 respectivamente):", med_people)
print("Media % de personas que aciertan (10, 11, 12, 13 y 14 respectivamente):", med_percent)
print("Media de premios unitarios (para 10, 11, 12, 13 y 14 respectivamente):", med_unitary)
```

```python
# Probabilidades teoricas
p = 1/3

probs = []
probs_full = []
med_per = np.array(med_percent) / 100
for k in range(10,15):
  probs.append(math.comb(14, k) * pow(p,k) * pow(1-p,14-k))

plt.figure(1)
axis_x = range(10, 15)
plt.bar(axis_x, probs)
plt.show()
print("\n", "Probabilidades teoricas de acertar: (10, 11, 12, 13 y 14 respectivamente) con p =", p)
print(np.array(probs))
print(np.around(1/np.array(probs)))
print("Probabilidad de que toque algo:", np.sum(probs), "= 1 /", round(1 / np.sum(probs)))
print("Pleno al 15 =", pow(p,14)/pow(4,2), "= 1 /", round(pow(4,2)/pow(p,14)), "\n")
plt.figure(1)
axis_x = range(10, 15)
plt.bar(axis_x, med_per, color="green")
plt.show()
print("\n Media % de personas que aciertan (10, 11, 12, 13 y 14 respectivamente) con p =", p)
print(med_per, "\n")

plt.figure(1)
axis_x = range(10, 15)
plt.plot(axis_x, np.array(abs(probs-med_per)), 'or')
plt.show()
print("\n Error (10, 11, 12, 13 y 14 respectivamente) con p = ", p)
print(np.array(abs(probs-med_per)), "\n")

for k in range(0,15):
  probs_full.append(math.comb(14, k) * pow(p,k) * pow(1-p,14-k))

plt.figure(1)
axis_x = range(0, 15)
plt.plot(axis_x, probs_full)
plt.show()
print("\n Funcion distribucion de probabilidad binomial de la Quiniela con p =", p)
print(probs_full, "\n")

MSE = np.square(np.subtract(probs, med_per)).mean()
print("MSE para p =", p, ":", MSE)
```

```python
# Optimizacion MSE
MSE = 1000
MSE_vec = []
p_good = 0
prob = 0
for p in np.arange(0, 0.5, 0.01):
  new_probs = []
  for k in range(10,15):
    prob = math.comb(14, k) * pow(p,k) * pow(1-p,14-k)
    new_probs.append(prob)
  MSE_new = np.square(np.subtract(new_probs, med_per)).mean()
  MSE_vec.append(MSE_new)
  if (MSE_new <= MSE):
    MSE = MSE_new
    p_good = p
print(MSE_vec)
print(p_good)


plt.figure(1)
axis_x = np.arange(0, 0.5, 0.01)
plt.plot(axis_x, MSE_vec, 'r')
plt.show()
print("La probabilidad que mejor se adapta al MSE es p =", p_good)
```
