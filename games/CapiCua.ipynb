# Probabilidad de Cap i Cua en una matricula de 4 numeros



# 0. Imports


```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random as ran
```

## 1. Probabilidad de primero y ultimo iguales

```python
# Probabilidad teórica
th = 1 / 10

# Numero de veces experimento
N = 10000

# Probabilidad calculada
hits = 0
for i in range(N):
  mat = np.random.randint(0, high=9, size=4)
  if mat[0] == mat[-1]:
    hits += 1

prob = hits / N
print("Probabilidad de último y primero igual TEORICA: ", th)
print("Probabilidad de último y primero igual estimada: ", prob)
```

## 2. Probabilidad de capicua
Se considera CapiCua cuando el primero y el último son iguales y el segundo y el penúltimo tambien

```python
# Probabilidad teórica
th_cc = 1 / 100

# Numero de veces experimento
N = 100000

# Probabilidad calculada
hits_cc = 0
for i in range(N):
  mat_cc = np.random.randint(0, high=9, size=4)
  if mat_cc[0] == mat_cc[-1] & mat_cc[1] == mat_cc[-2]:
    hits_cc += 1

prob_cc = hits_cc / N
print("Probabilidad de CapiCua TEORICA: ", th_cc)
print("Probabilidad de CapiCua estimada: ", prob_cc)
```
