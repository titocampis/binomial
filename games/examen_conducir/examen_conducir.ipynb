### Càlcul Probabilitat exacte *k* èxits

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Variables Definition
p = 1/3
n = 30
```

```python
k = 27

# Exact probability of a Binomial with a fixed k
result = math.comb(n, k) * pow(p,k) * pow(1-p,n-k)
print("Probabilitat d'encertat exactament", k, "partits de " + str(n) + ":", result, "~ 1 /", round(1/result))
```

### Càlcul Funció de Distribucí Binomial amb (p = 1/3, n = 30)

```python
# Calculation of distribution function of binomial with parameters (p,n)
results = []
print(f"Probabilitats nombre encerts amb p = {round(p,3)} i n = {n}")
for k in range(0,n+1):
  result = math.comb(n, k) * pow(p,k) * pow(1-p,n-k)
  results.append(result)
  print(f"P(X = {k}) = {result}")

print()
plt.figure(1)
axis_x = range(0, n+1)
plt.xlabel('Nombre encerts exactes')
plt.ylabel('Probabilitad')
plt.plot(axis_x, results, 'or')
plt.show()
```

### Càcul de la Probabilitat d'Aprovar

Cal tenir en compte que aprovar significa **encertar 27 o més preguntes**


```python
u = 27
accu = 0
probs_vector = []
print_vector = []
for k in range(u,n+1):
  print(f"P(X = {k}) = {results[k]}") # Using the previous calculated results
  probs_vector.append(results[k])
  print_vector.append(f"P(X = {k})")
  accu += results[k]

##################     Plot    ##################
print()
plt.figure(1)
axis_x = range(u,n+1)
plt.xlabel('Nombre encerts exactes')
plt.ylabel('Probabilitad')
plt.bar(axis_x, probs_vector)
plt.xticks(axis_x) # Set the tick labels to be the same as the values of axis_x

# Add text annotations above each column
for i, prob in zip(axis_x, probs_vector):
    if prob >= 0.00001: prob = round(prob,5)
    elif prob >= 0.000000001: prob = round(prob,9)
    else: prob = round(prob,15)
    plt.text(i, prob, str(prob), ha='center', va='bottom')
plt.show()
##################################################

print(f"\nP(X >= {u}) = {' + '.join(map(str, print_vector))}\n          = {' + '.join(map(str, probs_vector))}\n          = {accu}")
```

```python
all_probs = []
for k in range(0,31, 1):
  all_probs.append(math.comb(n, k) * pow(p,k) * pow(1-p,n-k))

plt.figure(1)
axis_x = range(0, 31)
plt.plot(axis_x, all_probs)
plt.show()
print("\n Funcion distribucion de probabilidad binomial del Examen de conducir con p =", p)
print(all_probs, "\n")
print("Comprobación:", sum(all_probs))
```

```python
acc_probs = []
for k in range(0,31, 1):
  if k == 0: acc_probs.append(math.comb(n, k) * pow(p,k) * pow(1-p,n-k))
  else:
    acc_probs.append(acc_probs[-1] + math.comb(n, k) * pow(p,k) * pow(1-p,n-k))
axis_x = range(0, 31)
plt.plot(axis_x, acc_probs)
plt.show()
print("\n Funcion de probabilidad binomial del Examen de Conducir")
```
