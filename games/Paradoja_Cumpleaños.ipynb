```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Paradoja del cumplea√±os, probabilidad de que en un grupo de n personas 2 o mas cumplan a√±os el mismo d√≠a
vec_prob = []
for n in range(2,50):
  vec_prob.append(1 - math.factorial(365) / (pow(365,n) * math.factorial(365-n)))

plt.figure(1)
axis_x = range(2, 50)
plt.plot(axis_x, vec_prob)
plt.show()

print_n = 49
print("Probabilidad de 2 o mas personas cumplan a√±o el mismo dia en un grupo de " + str(print_n) + " personas:", vec_prob[print_n-2], "= 1 /", round(1/vec_prob[print_n-2]))
```

```python
# Paradoja del cumplea√±os 2, probabilidad de que un grupo de n personas alg√∫n miembro cumpla a√±os el mismo d√≠a que yo
# 1 - Binomial(n,1/365) para k=0
vec_prob_2 = []
for n in range(2,1000):
  vec_prob_2.append(1 - pow(364/365,n))

plt.figure(1)
axis_x = range(2, 1000)
plt.plot(axis_x, vec_prob_2)
plt.show()

print_n = 1000
print("Probabilidad de que en un grupo de " + str(print_n) + " personas alg√∫n miembro cumpla a√±os el mismo d√≠a que yo:", vec_prob_2[print_n-3], "= 1 /", round(1/vec_prob_2[print_n-3]))
```
